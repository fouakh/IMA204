```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import procrustes, Delaunay
from sklearn.decomposition import PCA
import cv2
```

```python
class ActiveAppearanceModel:
    def __init__(self, shapes, images, texture_size):
        """
        Initialize the Active Appearance Model.
        """
        self.shapes = shapes
        self.images = images 
        self.texture_size = texture_size
        self.mean_shape = None
        self.textures = None

    def align_shapes(self):
        """
        Align all shapes to a reference shape using Procrustes analysis.
        """
        print("Aligning shapes with Procrustes...")
        reference_shape = np.mean(self.shapes, axis=0)
        aligned_shapes = []

        for shape in self.shapes:
            _, aligned_shape, _ = procrustes(reference_shape, shape)
            aligned_shapes.append(aligned_shape)

        self.shapes = [aligned * np.linalg.norm(shape - np.mean(shape, axis=0)) + 
                       np.mean(shape, axis=0) for aligned, shape in zip(aligned_shapes, self.shapes)]
        print("Shapes aligned successfully.")

    def plot_singular_values(self, pca_model, title):
        """
        Plot the singular values (explained variance) from a PCA.
        """
        plt.figure(figsize=(8, 4))
        plt.bar(range(1, len(pca_model.explained_variance_ratio_) + 1), pca_model.explained_variance_ratio_ * 100)
        plt.title(title)
        plt.xlabel("Principal Components")
        plt.ylabel("Explained Variance (%)")
        plt.show()

    def build_shape_model(self):
        """
        Build a statistical model for shapes using economical PCA with 95% explained variance.
        """
        print("Building shape model...")
        self.align_shapes()  # Align shapes before PCA
        shape_matrix = np.array([shape.flatten() for shape in self.shapes])
    
        # Keep components up to 95% variance explained
        self.shape_pca = PCA(n_components=0.95)
        self.shape_pca.fit(shape_matrix)
    
        self.shape_modes = self.shape_pca.components_
        self.shape_mean_vector = self.shape_pca.mean_
        self.mean_shape = self.shape_mean_vector.reshape(-1, 2)
    
        print(f"Shape model built with {len(self.shape_modes)} principal modes.")

        # Plot singular values
        self.plot_singular_values(self.shape_pca, "Singular Values of Shape PCA")

    def build_texture_model(self):
        """
        Build a statistical model for textures extracted from warped images using economical PCA.
        """
        print("Building texture model...")
    
        print("Warping textures to the mean shape...")
        self.textures = []
        for shape, image in zip(self.shapes, self.images):
            warped_texture = self.warp_texture(self.mean_shape, shape, image)
            self.textures.append(warped_texture.flatten())
        print("Textures warped successfully.")

        # Flatten the warped textures for PCA
        texture_matrix = np.array(self.textures)

        # Normalize textures
        self.normalized_textures = (texture_matrix - texture_matrix.mean(axis=1, keepdims=True)) / \
                                    (texture_matrix.std(axis=1, keepdims=True) + 1e-6)

        # Keep components up to 95% variance explained
        self.texture_pca = PCA(n_components=0.95)
        self.texture_pca.fit(self.normalized_textures)

        self.texture_modes = self.texture_pca.components_
        self.texture_mean_vector = self.texture_pca.mean_

        print(f"Texture model built with {len(self.texture_modes)} principal modes.")

        # Plot singular values
        self.plot_singular_values(self.texture_pca, "Singular Values of Texture PCA")

    def build_combined_model(self):
        """
        Build a combined statistical model for shape and texture using block-diagonal PCA.
        """
        print("Building combined appearance model...")

        # Perform PCA on combined data
        self.combined_pca = PCA(n_components=0.95)
        combined_data = np.hstack((
            self.shape_pca.transform(np.array([shape.flatten() for shape in self.shapes])),
            self.texture_pca.transform(np.array(self.normalized_textures))
        ))
        self.combined_pca.fit(combined_data)

        # Combined modes and projected mean
        self.combined_modes = self.combined_pca.components_
        self.combined_mean_vector = self.combined_pca.mean_

        # Recover full-size mean shape and texture
        shape_mean_full = self.shape_pca.inverse_transform(self.combined_mean_vector[:len(self.shape_modes)])
        texture_mean_full = self.texture_pca.inverse_transform(self.combined_mean_vector[len(self.shape_modes):])
    
        # Store the complete combined mean vector
        self.combined_mean_full_vector = np.hstack((shape_mean_full, texture_mean_full))

        print(f"Combined model built with {len(self.combined_modes)} principal modes.")

        # Plot singular values
        self.plot_singular_values(self.combined_pca, "Singular Values of Combined PCA")
        

    def warp_texture(self, mean_shape, target_shape, texture):
        """
        Warp the texture from the target shape to the mean shape using triangulation-based interpolation.
        """
        texture_image = texture.reshape(self.texture_size)
        
        # Perform Delaunay triangulation on the target shape
        triangulation = Delaunay(target_shape)
        
        warped_texture = np.zeros_like(texture_image)
        
        for simplex in triangulation.simplices:
            src_points = target_shape[simplex].astype(np.float32)
            dst_points = mean_shape[simplex].astype(np.float32)
            
            # Compute affine transform
            affine_transform = cv2.getAffineTransform(src_points, dst_points)
            
            # Mask for the current triangle
            mask = np.zeros(texture_image.shape, dtype=np.uint8)
            cv2.fillConvexPoly(mask, dst_points.astype(np.int32), 1)
            
            # Warp the triangle region
            warped_region = cv2.warpAffine(texture_image, affine_transform, texture_image.shape[::-1])
            
            # Blend the warped region into the result
            warped_texture[mask == 1] = warped_region[mask == 1]

        return warped_texture

    def generate_shape(self, shape_params):
        """
        Generate a shape from given shape parameters.
        """
        shape_vector = self.shape_mean_vector + np.dot(self.shape_modes.T, shape_params)
        return shape_vector.reshape(-1, 2)
    
    def generate_texture(self, texture_params):
        """
        Generate a texture warped to the mean shape from given texture parameters.
        """
        # Generate a texture vector
        generated_texture = self.texture_mean_vector + np.dot(self.texture_modes.T, texture_params)
        return generated_texture.reshape(self.texture_size)
    
    def generate_combined(self, params):
        """
        Generate a shape and texture from combined parameters.
        """
        # Compute combined appearance from parameters
        combined_projected = self.combined_mean_vector + np.dot(self.combined_modes.T, params)

        # Extract shape and texture components from the projected vector
        n_shape_components = len(self.shape_modes)
        shape_projected = combined_projected[:n_shape_components]
        texture_projected = combined_projected[n_shape_components:]

        # Recover the full shape and texture using inverse transforms
        shape = self.shape_pca.inverse_transform(shape_projected)
        texture = self.texture_pca.inverse_transform(texture_projected)

        return shape.reshape(-1, 2), texture.reshape(self.texture_size)
    
    def search(self, target_texture, training_images, parameter_displacements, sigma_displacement, max_iterations=50, tolerance=1e-3):
        """
        Perform Active Appearance Model search to fit the model to the target texture.
        """
        print("Starting AAM search...")

        # Precompute Jacobian matrix if not already computed
        self.precompute_jacobian(training_images, parameter_displacements, sigma_displacement)
        R = np.linalg.pinv(self.jacobian_r_p)  # Pseudo-inverse of the Jacobian

        # Initialize model parameters for combined shape, pose, and texture
        params = np.zeros(len(self.combined_modes))  # Shape and texture parameters
        t = np.zeros(4)  # Pose parameters [sx, sy, tx, ty]
        u = np.zeros(2)  # Texture intensity parameters [u1, u2]

        for iteration in range(max_iterations):
            # Generate current appearance with transformations
            shape, current_texture = self.generate_combined(params)
            shape = self.transform_shape(shape, t)
            current_texture = self.transform_texture(current_texture.flatten(), u)

            # Compute residual
            residual = target_texture - current_texture

            # Compute parameter update using precomputed R
            delta_combined_params = -np.dot(R, residual)
            delta_c = delta_combined_params[:len(params)]
            delta_t = delta_combined_params[len(params):len(params) + 4]
            delta_u = delta_combined_params[len(params) + 4:]

            # Update parameters
            params += delta_c
            t += delta_t
            u += delta_u

            # Check for convergence
            if np.linalg.norm(delta_combined_params) < tolerance:
                print(f"Converged after {iteration + 1} iterations.")
                break
        else:
            print("Max iterations reached without full convergence.")

        return params, t, u

    def precompute_jacobian(self, training_images, parameter_displacements, sigma_displacement):
        """
        Precompute the Jacobian matrix @r/@p using numerical differentiation.
        :param training_images: Set of training images for residual evaluation.
        :param parameter_displacements: List of displacements for each parameter.
        :param sigma_displacement: Standard deviations for parameter displacements.
        """
        num_params = len(self.combined_modes) + 4 + 2  # c (params), t (pose), u (intensity)
        self.jacobian_r_p = np.zeros((len(training_images[0].flatten()), num_params))

        print("Precomputing Jacobian using numerical differentiation...")

        for j in range(num_params):
            weighted_residuals_sum = np.zeros_like(training_images[0].flatten())
            weight_sum = 0

            # Iterate over different displacements
            for displacement in parameter_displacements:
                delta = np.zeros(num_params)
                delta[j] = displacement

                # Apply displacement and generate new appearance
                perturbed_params = np.zeros(len(self.combined_modes)) if j < len(self.combined_modes) else None
                perturbed_t = np.zeros(4) if len(self.combined_modes) <= j < len(self.combined_modes) + 4 else None
                perturbed_u = np.zeros(2) if j >= len(self.combined_modes) + 4 else None

                if perturbed_params is not None:
                    perturbed_params[j] = displacement
                elif perturbed_t is not None:
                    perturbed_t[j - len(self.combined_modes)] = displacement
                elif perturbed_u is not None:
                    perturbed_u[j - len(self.combined_modes) - 4] = displacement

                shape, perturbed_texture = self.generate_combined(perturbed_params)
                shape = self.transform_shape(shape, perturbed_t) if perturbed_t is not None else shape
                perturbed_texture = self.transform_texture(perturbed_texture.flatten(), perturbed_u) if perturbed_u is not None else perturbed_texture

                # Compute residual difference
                residual = training_images[0].flatten() - perturbed_texture.flatten()

                # Compute weight
                weight = np.exp(-displacement ** 2 / (2 * sigma_displacement ** 2)) / displacement
                weighted_residuals_sum += weight * residual
                weight_sum += weight

            # Compute final average for the Jacobian entry
            self.jacobian_r_p[:, j] = weighted_residuals_sum / weight_sum

        print("Jacobian precomputation completed.")
        
        
    def transform_shape(self, shape, t):
        """
        Apply pose transformation to shape.
        """
        sx, sy, tx, ty = t
        s = np.sqrt((sx + 1) ** 2 + sy ** 2) + 1e-6
        theta = np.arctan2(sy, sx + 1)

        rotation_scaling = np.array([
            [s * np.cos(theta), -s * np.sin(theta)],
            [s * np.sin(theta),  s * np.cos(theta)]
        ])

        transformed_shape = np.dot(shape, rotation_scaling.T) + np.array([tx, ty])
        return transformed_shape

    def transform_texture(self, texture, u):
        """
        Apply texture intensity transformation.
        """
        u1, u2 = u
        return (1 + u1) * texture + u
```

```python
# Load dataset from Data_FEI.npz
Working_directory = "./"
with np.load(Working_directory + 'Data_FEI.npz') as data:
    Images = data['Images_FEI']  # List of images
    Landmarks = data['Landmarks_FEI']  # Original landmarks
    Y = data['Emotions_FEI']  # Emotion classes (0 for neutral, 1 for happy)
    Names = data['Names_FEI']

N, M = Landmarks.shape  # Number of subjects, dimension of landmarks
dim = 2
M = int(M / dim)  # Number of landmark points per subject
print(f'Number of subjects: {N}, Number of landmarks: {M}')

# Reshape landmarks and images
shapes = [Landmarks[i].reshape(M, 2) for i in range(N)]
texture_size = (300, 250) 
textures = [Images[i].reshape(texture_size) for i in range(N)]
```

```python
# Create and build the model
aam = ActiveAppearanceModel(shapes, textures, texture_size)
```

```python
aam.build_shape_model()
```

```python
# Plot the mean shape
plt.figure(figsize=(15, 4))
plt.subplot(1, 5, 1)
mean_shape = aam.mean_shape
plt.scatter(mean_shape[:, 0], mean_shape[:, 1], color='red')  # Flip the y-axis values
plt.title("Mean Shape")
plt.axis('equal')
plt.gca().invert_yaxis()

# Number of modes to visualize and parameter variation
num_modes = 4
param_variation = 100.0

# Plot variations along the first few modes
for i in range(num_modes):
    # Positive variation
    b = np.zeros(len(aam.shape_modes))
    b[i] = param_variation

    # Generate shape with positive variation
    shape_pos = aam.generate_shape(b)

    plt.subplot(1, num_modes + 1, i + 2)
    plt.scatter(shape_pos[:, 0], shape_pos[:, 1], color='blue')  # Flip y-axis values
    plt.title(f"Mode {i + 1} (+{param_variation}σ)")
    plt.axis('equal')
    plt.gca().invert_yaxis()

plt.show()
```

```python
aam.build_texture_model()
```

```python
# Plot the mean texture
plt.figure(figsize=(15, 4))
plt.subplot(1, 5, 1)

# Plot the mean texture
mean_texture = aam.texture_mean_vector.reshape(aam.texture_size)
plt.imshow(mean_texture, cmap='gray')
plt.title("Mean Texture")
plt.axis('equal')

# Number of modes to visualize and parameter variation
num_modes = 4
param_variation = 100.0

# Plot variations along the first few modes
for i in range(num_modes):
    # Positive variation
    b = np.zeros(len(aam.texture_modes))
    b[i] = param_variation

    # Generate texture with positive variation
    texture_pos = aam.generate_texture(b)

    plt.subplot(1, num_modes + 1, i + 2)
    plt.imshow(texture_pos, cmap='gray')
    plt.title(f"Mode {i + 1} (+{param_variation}σ)")
    plt.axis('equal')

plt.show()
```

```python
aam.build_combined_model()
```

```python
plt.figure(figsize=(15, 4))
plt.subplot(1, 5, 1)

# Extract mean shape and texture
mean_combined_shape = aam.combined_mean_full_vector[:len(aam.shape_mean_vector)].reshape(-1, 2)
mean_combined_texture = aam.combined_mean_full_vector[len(aam.shape_mean_vector):].reshape(aam.texture_size)

# Plot mean texture
plt.imshow(mean_combined_texture, cmap='gray')
plt.title("Mean Appearance")
plt.axis('equal')

# Plot variations along the first few combined modes
num_modes = 4
param_variation = 100.0

for i in range(num_modes):
    # Positive variation
    b = np.zeros(len(aam.combined_modes))
    b[i] = param_variation

    # Generate appearance with positive variation
    _ , texture_pos = aam.generate_combined(b)

    plt.subplot(1, num_modes + 1, i + 2)
    plt.imshow(texture_pos, cmap='gray')
    plt.title(f"Mode {i + 1} (+{param_variation}σ)")
    plt.axis('equal')

plt.show()
```
